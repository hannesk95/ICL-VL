```python
# Import required libraries
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Load the NIfTI image
file_path = 'data/glioma/glioma_four_sequences/UCSF-PDGM-0004_T1c_bias.nii.gz'
img = nib.load(file_path)
img_data = img.get_fdata()

# Get center slices for axial, sagittal, and coronal views
axial_slice = img_data.shape[2] // 2
sagittal_slice = img_data.shape[0] // 2
coronal_slice = img_data.shape[1] // 2

# Plotting the 3 views
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Axial view (view from top)
axes[0].imshow(img_data[:, :, axial_slice], cmap="gray")
axes[0].set_title(f'Axial Slice {axial_slice}')

# Sagittal view (side view)
axes[1].imshow(img_data[sagittal_slice, :, :].T, cmap="gray", origin="lower")
axes[1].set_title(f'Sagittal Slice {sagittal_slice}')

# Coronal view (front view)
axes[2].imshow(img_data[:, coronal_slice, :].T, cmap="gray", origin="lower")
axes[2].set_title(f'Coronal Slice {coronal_slice}')

plt.tight_layout()
plt.show()
```
